```python
a = np.linspace(0, 100, 1000)
b = np.sin(a) + np.sin(3*a) + np.sin(5*a)
plt.plot(b)
```

```python
# Using Numpy
Y = np.fft.fft(b)
plt.plot(np.abs(Y))
```
