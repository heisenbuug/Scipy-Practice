```python
from scipy import signal as sg
import numpy as np
```

```python
a = b = np.arange(25).reshape((5, 5))
```

```python
a
```

```python
b
```

### 2D Convolution

```python
sg.convolve2d(a, b)
```

### 2D Corelation

```python
sg.correlate2d(a, b)
```

### For 3D Matrix

```python
a = b = np.arange(125).reshape((5, 5, 5))
```

```python
sg.convolve(a, b)
```

```python
sg.correlate(a, b)
```

### Fast Fourier Transform

```python
sg.fftconvolve(a, b)
```
