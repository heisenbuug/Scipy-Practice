```python
%matplotlib inline
import scipy.signal as sg
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2, 4000)
```

```python
signal_space = 2 * np.pi * 3 * t
```

### Sawtooth Wave

```python
plt.plot(t, sg.sawtooth(signal_space))
plt.title('Sawtooth Signal')
```

### Square Wave

```python
plt.plot(t, sg.square(signal_space))
plt.title('Square Wave')
```

### Modulation Of Waves

```python
# Pulse Width Modulation
plt.subplots_adjust(wspace=0.5, hspace=0.5)
signal_space = 2 * np.pi * t
sig = np.sin(signal_space)
pwm = sg.square(30 * signal_space, duty=(sig+1)/2)

plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.title('Sine Wave')

plt.subplot(2, 1, 2)
plt.plot(t, pwm)
plt.title('PWM Wave')
```

### Gaussian Pulse

```python
t = np.linspace(-1, 1, 2*100)
```

```python
# Center Frequency = 4
i, q, e = sg.gausspulse(t, fc=4, retquad=True, retenv=True)
plt.plot(t, i, '-')
```

```python
plt.plot(t, q, '--')
```

```python
plt.plot(t, e, '-.')
```

```python
plt.plot(t, i, '-') # Blue
plt.plot(t, q, '--') # Orange
plt.plot(t, e, '-.') # Green
plt.show()
```

### Chirp Pulse

```python
t = np.linspace(0, 10, 501)
```

```python
# f0: frequency at time t
# f1: frequency at time t1
w = sg.chirp(t, f0=1, f1=10, t1=20, method='linear')
```

```python
plt.plot(t, w)
plt.title('Linear Chirp Pulse')
plt.xlabel('t (sec)')
```
